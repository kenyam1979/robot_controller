```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pykalman import KalmanFilter
```

```python
raw = pd.read_csv('data2.csv', names=['left', 'right'])
```

```python
d = raw[250:500].reset_index()
```

```python
d
```

```python
len(d)
```

```python
fig, axs = plt.subplots(1, 1)
axs.plot(d.left)
plt.show()
```

```python
N_STATE = 20
initial_state_mean = np.zeros(N_STATE)
initial_state_covariance = np.ones((N_STATE, N_STATE))
F = np.diag(np.ones(N_STATE))
G = np.zeros((1, N_STATE))
G[0, 0] =1.0
```

```python
# kf = KalmanFilter(
#     n_dim_obs=,
#     n_dim_state=N_STATE,
#     initial_state_mean=initial_state_mean, 
#     initial_state_covariance=initial_state_covariance
#     transition_matrices=F,
#     observation_matrices=H,
#     observation_covariance=0.04,
#     transition_covariance=XXX)
```

```python
kf = KalmanFilter(
    n_dim_obs=1,
    n_dim_state=N_STATE,
    initial_state_mean=initial_state_mean, 
    initial_state_covariance=initial_state_covariance)
kf = kf.em(d.left)
```

```python
kf.observation_matrices
```

```python
kf.observation_covariance
```

```python
kf.transition_matrices
```

```python
kf.transition_covariance
```

```python
pred = [0] * len(d)
current_state_mean = initial_state_mean
current_state_covariance = initial_state_covariance
for i, obs in enumerate(d.left):
    current_state_mean, current_state_covariance = kf.filter_update(
        current_state_mean, 
        current_state_covariance, 
        observation=obs)
    pred[i] = kf.observation_matrices.dot(current_state_mean)
```

```python
fig, axs = plt.subplots(1, 1)
axs.plot(d.left)
axs.plot(pred)
plt.show()
```
